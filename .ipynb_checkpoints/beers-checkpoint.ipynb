```python
data = data.drop(['BeerID','Name', 'URL', 'Style'], axis=1)
```

```python
data.head()
```

```python
data = data.fillna(data.mean())
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['SugarScale', 'BrewMethod'])
```

```python
len(data['PrimingMethod'].unique())
```

```python
len(data['PrimingAmount'].unique())
```

```python
73861 / len(data['PrimingAmount'].unique())
```

```python
data.head()
```

```python
data = data.drop(['PrimingAmount','PrimingMethod'], axis=1)
```

```python
data = data.sample(frac=1)
cnt = data.shape[0]
index1 = int(np.ceil(cnt * 0.7))
index2 = int(np.ceil(cnt * 0.85))
data_train = data.iloc[1:index1]
data_eval = data.iloc[index1+1:index2]
data_test = data.iloc[index2+1:]
print('Sizes:\nTraining set - %d\nEval set       - %d\nTest set     - %d' % (data_train.shape[0], data_eval.shape[0], data_test.shape[0]))
```

```python
Y_train = data_train['StyleID'].values
# Y_train = np.reshape(Y_train, (1, Y_train.shape[0]))
Y_test = data_test['StyleID'].values 
# Y_test = np.reshape(Y_test, (1, Y_test.shape[0]))
Y_eval = data_eval['StyleID'].values 
# Y_eval = np.reshape(Y_eval, (1, Y_eval.shape[0]))

data_train = data_train.drop(['StyleID'], axis=1)
data_test = data_test.drop(['StyleID'], axis=1)
data_eval = data_eval.drop(['StyleID'], axis=1)


X_train = data_train.as_matrix()
X_test = data_test.as_matrix()
X_eval = data_eval.as_matrix()
```

```python
np.save('Y_train',Y_train)
np.save('Y_test',Y_test)
np.save('Y_eval',Y_eval)

np.save('X_train',X_train)
np.save('X_test',X_test)
np.save('X_eval',X_eval)
```
